<h2 style="padding: 5px 0px; color:#0b0265; font-weight: bold; font-family: Cursive">
Import Libraries</h2>


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

<h2 style="padding: 5px 0px; color:#0b0265; font-weight: bold; font-family: Cursive">
Loading Data</h2>

```python
iris = load_iris()
X = iris.data
```

To predict the optimum number of clusters in the 'Iris' dataset, we can use the
- Elbow method 
- Silhouette analysis.

<h2 style="padding: 5px 0px; color:#0b0265; font-weight: bold; font-family: Cursive">
Elbow Methods</h2>

The elbow method works by plotting the within-cluster sum of squares (WCSS) against the number of clusters. The WCSS measures the distance between each point and its corresponding cluster center. As the number of clusters increases, the WCSS decreases, but at a certain point, the rate of decrease slows down, resulting in a bend in the plot that looks like an elbow. This bend in the plot represents the optimal number of clusters, as adding more clusters after this point does not result in a significant decrease in WCSS.

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

<h2 style="padding: 5px 0px; color:#0b0265; font-weight: bold; font-family: Cursive">
Silhouette Analysis</h2>
The silhouette analysis measures the similarity of each data point to its own cluster compared to other clusters. The silhouette score ranges from -1 to 1, where a high score indicates that the point is well-matched to its own cluster and poorly-matched to neighboring clusters. The average silhouette score for all data points is calculated for different numbers of clusters, and the number of clusters with the highest average silhouette score is considered the optimal number of clusters.

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []
for n_cluster in range(2, 11):
    silhouette_scores.append(silhouette_score(X, KMeans(n_clusters=n_cluster).fit_predict(X)))
    
plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

<h2 style="padding: 5px 0px; color:#0b0265; font-weight: bold; font-family: Cursive">
Model</h2>

```python
#model
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

<h2 style="padding: 5px 0px; color:#0b0265; font-weight: bold; font-family: Cursive">
Visualization</h2>

```python
# Create a figure object with a larger size
fig = plt.figure(figsize=(10, 7))
colors = ['Blue', 'green', 'red']
# Plot the scatter plot with larger marker size
for i in range(3):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], s=200, c=colors[i], label='Iris-'+iris.target_names[i])

# Plot the centroids with larger marker size
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='black', label='Centroids')

# Set plot title, axis labels and legend
plt.title('Iris Clusters')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend()

# Show the plot
plt.show()
```
